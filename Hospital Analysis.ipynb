### Number of Patients with different Length of Stay

```python
more_than_7 = 0
more_than_15 = 0
less_than_1 = 0
average_LOS = 0
total = 0
date1 = datetime.strptime('13-01-01 00:00:00', date_format)
date2 = datetime.strptime('13-12-31 23:59:59', date_format)
for row,column in db1.iterrows():
    admission_day = column['Admission date and time']
    if admission_day >= date1 and admission_day < date2:
        total+=1
        if column['LOS'].days >= 7 and column['LOS'].days < 15:
            more_than_7+=1
        elif column['LOS'].days >15:
            more_than_15+=1
        elif column['LOS'].days == 0:
            less_than_1+=1
        elif column['LOS'].days > 0 and column['LOS'].days < 7:
            average_LOS+=1
print("{} patients stayed between 7 and 15 days ".format(more_than_7))
print("{} patients stayed for more than 15 days".format(more_than_15))
print("{} patients stayed for less than 1 day".format(less_than_1))
print("{} patients stayed betwwen 0 and 7 days".format(average_LOS))
print("{} total number of patients admitted".format(total))
```

```python
adm_num = []
dis_num = []
diff = []
date = []
ad_Stroke = []
ad_Neur = []
ad_Ur = []
ad_Card = []
paed_Surg = []
#start_time = datetime.strptime('00:00:00', '%H:%M:%S')
#end_time = datetime.strptime('23:59:59', '%H:%M:%S')
hours = 23
minutes = 59
seconds = 59
add_hours = timedelta(hours = hours, minutes = minutes, seconds = seconds)
for row,column in dataset_date.iterrows():
    date_start = column['Date 2013']
    date_end = column['Date 2013'] + add_hours
    adm = db1[(db1['Admission date and time']>=date_start) & (db1['Admission date and time']<date_end)]
    dis = db1[(db1['Discharge Date and Time']>=date_start) & (db1['Discharge Date and Time']<date_end)]
    ad_S = db1[(db1['Admission date and time']>=date_start) & (db1['Admission date and time']<date_end) & (db1['Specialty'] == 'Adult Stroke')]
    ad_N = db1[(db1['Admission date and time']>=date_start) & (db1['Admission date and time']<date_end) & (db1['Specialty'] == 'Adult Neurology')]
    ad_U = db1[(db1['Admission date and time']>=date_start) & (db1['Admission date and time']<date_end) & (db1['Specialty'] == 'Adult Urology')]
    ad_C = db1[(db1['Admission date and time']>=date_start) & (db1['Admission date and time']<date_end) & (db1['Specialty'] == 'Adult Cardiology')]
    pae_S = db1[(db1['Admission date and time']>=date_start) & (db1['Admission date and time']<date_end) & (db1['Specialty'] == 'Paediatric Surgery')]
    adm_num.append(str(len(adm)))
    dis_num.append(str(len(dis)))
    difference = len(adm) - len(dis)
    diff.append(difference)
    date.append(date_start)
    ad_Stroke.append(str(len(ad_S)))
    ad_Neur.append(str(len(ad_N)))
    ad_Ur.append(str(len(ad_U)))
    ad_Card.append(str(len(ad_C)))
    paed_Surg.append(str(len(pae_S)))
```

```python
LOS_Group = []
total = 0
date1 = datetime.strptime('12-01-01 00:00:00', date_format)
date2 = datetime.strptime('16-01-31 23:59:59', date_format)
for row,column in db1.iterrows():
    admission_day = column['Admission date and time']
    if admission_day >= date1 and admission_day < date2:
        total+=1
        if column['LOS'].days >= 7 and column['LOS'].days < 15:
            LOS_Group.append('7-15')
        elif column['LOS'].days >=15:
            LOS_Group.append('15+')
        elif column['LOS'].days == 0:
            LOS_Group.append('0-')
        elif column['LOS'].days > 0 and column['LOS'].days < 7:
            LOS_Group.append('0-7')
```
